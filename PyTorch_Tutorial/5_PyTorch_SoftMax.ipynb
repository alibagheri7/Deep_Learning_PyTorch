cross entropy : the better the prediction the lower the lossimage.png

```python
def cross_entropy(actual, predicted):
  loss = - np.sum(actual * np.log(predicted))
  return loss
```

```python
from numpy.core.numeric import cross
Y = np.array([1,0,0])

Y_pred_good = np.array([0.7,.02,.1])
Y_pred_bad  = np.array([0.1,0.3,0.6])

l1 = cross_entropy(Y,Y_pred_good)
l2 = cross_entropy(Y,Y_pred_bad)
print(l1 , l2)
```

```python
import torch
import torch.nn as nn
import numpy as np
```

```python
loss = nn.CrossEntropyLoss()
```

```python
Y = torch.tensor([0])
# for prediction raw value and not the tranformed and normalised ones
Y_pred_good = torch.tensor([[2,1,.1]])
Y_pred_bad = torch.tensor([[0.5,2,.3]])

l1 = loss(Y_pred_good, Y)
l2 = loss(Y_pred_bad, Y)

print(l1.item())
print(l2.item())
```

```python
# 3 samples
# first correct class is 2, the second is 0 and the third is 1
Y = torch.tensor([2,0,1])
# for prediction raw value and not the tranformed and normalised ones
Y_pred_good = torch.tensor(
    [[0.1, 0.2, 3.9], # predict class 2
    [1.2, 0.1, 0.3], # predict class 0
    [0.3, 2.2, 0.2]]) # predict class 1

Y_pred_bad = torch.tensor(
    [[0.9, 0.2, 0.1],
    [0.1, 0.3, 1.5],
    [1.2, 0.2, 0.5]])

l1 = loss(Y_pred_good, Y)
l2 = loss(Y_pred_bad, Y)

print(l1.item())
print(l2.item())
```
